```python
from sklearn.datasets import make_moons
import numpy as np
import matplotlib.pyplot as plt

from nn import Sequential, Input, Dense, Tanh
```

```python
X, Y = make_moons(noise=0.1)
Y = np.expand_dims(Y, 1)
plt.scatter(X.T[0], X.T[1], c=Y)
plt.show()
```

```python
X = np.reshape(X, (100, 2, 1))
Y = np.reshape(Y, (100, 1, 1))
X.shape, Y.shape
```

```python
model = Sequential([
    Input(X.shape[1]),
    Dense(4),
    Tanh(),
    Dense(4),
    Tanh(),
    Dense(Y.shape[1]),
    Tanh(),
])
model.compile()
```

```python
model.fit(X, Y, epochs=10000, learning_rate=0.01)
```

```python
x1 = np.random.rand(1000, 1) * 1.75 - 0.5
x2 = np.random.rand(1000, 1) * 3.00 - 1.0
x = np.vstack([x2, x1])
x = np.reshape(x, (1000, 2, 1))
# y = np.reshape(Y, (100, 1))
y_pred = np.reshape(model.predict(x), (1000, 1))
x = np.reshape(x, (1000, 2))
plt.scatter(x.T[0], x.T[1], c=y_pred)
ax = plt.gca()
ax.set_xlim([-1, 2])
ax.set_ylim([-0.5, 1.25])
plt.show()
# Y[:5], model.predict(X[:5])
```
